#### Hypothesis

Null Hypothesis(H0): Any kind of statement is called as NUll Hypothesis unless it has been proven.
eg: (average(radius of straw)=0.05)

Alternate Hypothesis (H1):If average is lesser or greater than the value we have provided. 
eg:-radius of straw is greater or lesser than the actual value.

In real life, its nearly impossible to proof Null Hypothesis.

Sometimes, finding the alternative hypothesis is easier than null hypothesis.

Two Tail Test: In alterate hypothesis, we are trying to test the lesser than scenario as well as the greater than scenario. This type of testing is called Two tail test.


To approve or disapprove hypothesis, Critical region plays a vital role:

A Criitical Region is a region with respect to our dataset which creates a value that we reject inside a null hypothesis.

Suppose:

null hypothesis:
average=0.05

Alternate hypothesis:
_a_<0.05<_b_

any value more than 15000 is critical region.

here 15000-30000 is a critical region.

If critical region is present in both sides, we can perform the two tail test.

Type 1 Error is also called as False Positive. (Null hypothesis is correct, but we have proven its not correct).
